```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
canada = data[data.country == 'Canada']
china = data[data.country == 'China']
india = data[data.country == 'India']
us = data[data.country ==  'United States']
brazil = data[data.country == 'Brazil']
russia = data[data.country == 'Russia']
japan = data[data.country == 'Japan']
```

```python
plt.plot(us.year, us.population / 10**6, 'r.--' ,linewidth=1.5)
plt.plot(india.year, india.population / 10**6, 'g.--' ,linewidth=2)
plt.plot(canada.year, canada.population / 10**6, 'b.--' ,linewidth=1)
plt.plot(china.year, china.population / 10**6, 'y.--' ,linewidth=2)
plt.plot(brazil.year, brazil.population / 10**6, '.--' ,linewidth=2)
plt.plot(russia.year, russia.population / 10**6, '.--' ,linewidth=2)

plt.legend(['United states', 'India' , 'Canada', 'China', 'Brazil', 'Russia'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
